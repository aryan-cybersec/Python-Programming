**Q1) Find a factorial of given number.**

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        result = 1
        for i in range(2, n+1):
            result *= i
        return result   # ✅ Moved outside the loop

num = int(input("Enter a Number: "))
print("Factorial of", num, "is", factorial(num))
```

**Q2) Find whether the given number is Armstrong number.**

```python
def is_armstrong(num):

    digits = str(num)
    power = len(digits)
    

    total = sum(int(digit) ** power for digit in digits)
    

    return total == num

num = int(input("Enter a number: "))
if is_armstrong(num):
    print(num, "is an Armstrong number")
else:
    print(num, "is not an Armstrong number")
```

**Q3) Print Fibonacci series up to given term.**

```python
def fibonacci(n):
    a,b=0,1
    series=[]
    for _ in range(n):
        series.append(a)
        a,b=b,a+b
    return series
num=int(input("Enter the number of terms: "))
print("Fibonacci serires up to",num,"terms is:",fibonacci(num))
```

**Q4) Write a program to find if given number is prime number or not.**

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):  
         if n % i == 0:
            return False
    return True

num = int(input("Enter a number: "))
if is_prime(num):
    print(num, "is a Prime number")
else:
    print(num, "is not a Prime number")
```

**Q5) Check whether given number is palindrome or not.**

```python
def palindrome(num):
   
    s = str(num)
   
    return s == s[::-1]
num = int(input("Enter a number: "))
if palindrome(num):
    print(num, "is a Palindrome number")
else:
    print(num, "is not a Palindrome number")
```

**Q6) Write a program to print sum of digits.**

```python
def sum_of_digits(n):
    total = 0
    while n > 0:
        digit = n % 10      
        total += digit      
        n //= 10            
    return total


num = int(input("Enter a number: "))
print("Sum of digits of", num, "is", sum_of_digits(num))
```

**Q7) Count and print all numbers divisible by 5 or 7 between 1 to 100.**

```python
def divisible():
    num=[]
    for i in range (1,101):
        if i%5==0 or i%7==0:
            num.append(i)
    return num
result=divisible()
print("Number divisible by 5 or 7 between 1 to 101 are:")
print(result)
print("Total count:",len(result))
```

**Q9) Print all prime numbers between 1 and 100.**

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):  
        if n % i == 0:
            return False
    return True


primes = [i for i in range(1, 101) if is_prime(i)]
print("Prime numbers between 1 and 100 are:")
print(primes)
```
