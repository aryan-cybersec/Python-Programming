*   5.Check whether the quadratic equation has real roots or imaginary roots. Display the roots.

```python
import cmath  # cmath handles complex numbers

# Quadratic equation: ax^2 + bx + c = 0
a = float(input("Enter coefficient a: "))
b = float(input("Enter coefficient b: "))
c = float(input("Enter coefficient c: "))

# Calculate discriminant
d = b**2 - 4*a*c

# Find roots using quadratic formula
root1 = (-b + cmath.sqrt(d)) / (2*a)
root2 = (-b - cmath.sqrt(d)) / (2*a)

# Check nature of roots
if d > 0:
    print("The equation has real and distinct roots.")
elif d == 0:
    print("The equation has real and equal roots.")
else:
    print("The equation has imaginary (complex) roots.")


print("Root 1:", root1)
print("Root 2:", root2)
```

* 7.Write a program which takes any date as input and display next date of the calendar e.g.
I/P: day=20 month=9 year=2005
O/P: day=21 month=9 year 2005

```python
import datetime

# Input date
day = int(input("Enter day: "))
month = int(input("Enter month: "))
year = int(input("Enter year: "))

# Create a date object
current_date = datetime.date(year, month, day)

# Add one day
next_date = current_date + datetime.timedelta(days=1)

# Display result
print("Next Date is: day =", next_date.day, "month =", next_date.month, "year =", next_date.year)
```
