Q2) Count total number of vowels in a given string.

Q4) WAP to enter a string and a substring. You have to print the number of times 
that the substring occurs in the given string. String traversal will take place from 
left to right, not from right to left.  
Sample Input  
ABCDCDC  
CDC  
Sample Output  
2 

```python
#Solution:
string = input("Enter the main string: ")
substring = input("Enter the substring: ")
count = 0 
length = len(substring)
for i in range (len(string)- length + 1 ):
    if string[i:i+ length] = substring:
        count = count + 1
print(count)
```

Q5) Given a string containing both upper and lower-case alphabets. Write a Python 
program to count the number of occurrences of each alphabet (case 
insensitive) and display the same.  
Sample Input  
ABaBCbGc  
Sample Output  
2A  
3B  
2C  
1G

```python
#Solution:
s = input("Enter a string: ")
s = s.upper()
checked = ""
for ch in s :
    if ch.isalpha() and ch not in checked:
        count = s.count(ch)
        print(count, ch)
        checked +=ch
```

6. Program to count the number of unique words in a given sentence using sets. 

```python
#Solution:
sentence = input("Enter a sentence: ")

words = sentence.split()
unique_words = set(words)
print("Unique words are: ", unique_words)
print("Number of unique words:", len(unique_words))


```

7. Create 2 sets s1 and s2 of n fruits each by taking input from user and find:  
a. Fruits which are in both sets s1 and s2  
b. Fruits only in s1 but not in s2  
c. Count of all fruits from s1 and s2  

```python
#Solution:
n = int(input("Enter a number of fruits in each set: "))
s1= set()
s2 = set()
print("Enter fruits for set s1: ")
for i in  range(n):
    fruit = input()
    s1.add(fruit)
print("Enter fruits for set s2: ")
for i in range(n):
    fruit = input()
    s2.add(fruit)

common_fruits = s1 & s2
only_s1 = s1- s2
all_fruits = s1 | s2

print("Fruits in both s1 and s2 :", common_fruits)
print("Fruits only in s1: ", only_s1)
print("Total count of fruits: ", len(all_fruits))
```

8. Take two sets and apply various set operations on them:  
S1 = {Red, yellow, orange, blue}  
S2 = {violet, blue, purple}

```python
#Solution
S1 = {"Red", "yellow", "orange", "blue"}  
S2 = {"violet", "blue", "purple"}
print("S1: ", S1)
print("S2: ", S2)

print("Union:", S1 | S2)

print("Intersection: ", S1 & S2)

print("S1-S2", S1- S2)
print("S2- S1", S2- S1)
print("Symmetric Difference:", S1 ^ S2)


```
