1. Scan n values in range 0-3 and print the number of times each value has 
occurred. 

```python
#Solution:
n = int(input("Enter number of values: "))
count = [0,0,0,0]
for i in range(n):
    value = int(input("Enter value (0-3): "))
    if 0 <=value <= 3:
        count[value] += 1
for i in range(4):
    print(f"{i}occurred {count[i]} times")
```

2. Create a tuple to store n numeric values and find the average of all values. 

```python
#Solution:
n = int(input("Enter number of elements: "))
numbers = []
for i in range(n):
    numbers.append(int(input("Enter number: ")))
t = tuple(numbers)
average = sum(t) / len(t)
print("Tuple: ", t)
print("Average: ", average)
```

3. WAP to input a list of scores for N students in a list data type. Find the score of

Sample Input  
N = 5  
Scores= 2 3 6 6 5  
Sample output  
5  
Note: Given list is [2, 3, 6, 6, 5]. The maximum score is 6, second maximum is 

 Hence, we print 5 as the runner-up score. 

```python
#Solution:
N = int(input("Enter number of students: "))
scores = list(map(int, input("Enter scores").split()))
unique_scores = list(set(scores))
unique_scores.sort()
runner_up = unique_scores[-2]
print("Runner- up score :", runner_up)
```

4. Create a dictionary of n persons where key is name and value is city.   
a. Display all names  
b. Display all city names  
c. Display student names and cities of all students.  
d. Count the number of students in each city. 

```python
#Solution:
n = int(input("Enter number of students: "))
students = {}
for i in range(n):
    name = input("Enter name: ")
    city = input("Enter city: ")
    students[name]= city
print("Names:")
for name in students.keys():
    print(name)
print("Cities: ")
for city in students.values():
    print(city)
print("Students and cities: ")
for name, city in students.items():
    print(name, "- ", city)
city_count = 0
for city in students.values():
    city_count[city_count.get(city, 0)+ 1]
print("Students per city: ")
for city, count in city_count.items():
    print(city, ":", count)
```
