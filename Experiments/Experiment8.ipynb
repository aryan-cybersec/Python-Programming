Question 1. Add few names, one name in each row, in “name.txt file”.
a. Count no of names
b. Count all names starting with vowel
c. Find longest nam

```python
# open the file and read names
with open("name.txt", "r") as file:
    names = file.read().splitlines()   # removes newline characters

# a) Count total number of names
total_names = len(names)

# b) Count names starting with vowel
vowels = ('A', 'E', 'I', 'O', 'U', 'a', 'e', 'i', 'o', 'u')
vowel_names = [name for name in names if name.startswith(vowels)]
vowel_count = len(vowel_names)

# c) Find longest name
longest_name = max(names, key=len)

# print results
print("Total names:", total_names)
print("Names starting with vowel:", vowel_count)
print("Longest name:", longest_name)
```

Question2: 2. Store integers in a file.
a. Find the max number
b. Find average of all numbers
c. Count number of numbers greater than 100

```python
#Solution: 
# Program to:
# a) Find maximum number
# b) Find average of all numbers
# c) Count numbers greater than 100

# Open the file in read mode
with open("numbers.txt", "r") as file:
    
    # Read all lines from the file
    lines = file.read().splitlines()
    
    # Convert each line (string) into integer
    numbers = [int(num) for num in lines]

# a) Find the maximum number
max_number = max(numbers)

# b) Find the average
average = sum(numbers) / len(numbers)

# c) Count numbers greater than 100
count_gt_100 = 0
for num in numbers:
    if num > 100:
        count_gt_100 += 1

# Print results
print("Numbers in file:", numbers)
print("Maximum number:", max_number)
print("Average of numbers:", average)
print("Count of numbers greater than 100:", count_gt_100)
```

#Question3: Assume a file city.txt with details of 5 cities in given format (cityname 
population(in lakhs) area(in sq KM) ):
Example:
Dehradun 5.78 308.20
Delhi 190 1484
……………
Open file city.txt and read to:
a. Display details of all cities
b. Display city names with population more than 10Lakhs
c. Display sum of areas of all cities

```python
# Program to read city.txt and perform operations

# Open the file in read mode
with open("city.txt", "r") as file:
    
    # Read all lines from file
    lines = file.read().splitlines()

# Create empty lists to store data
cities = []
populations = []
areas = []

# Split each line and store values separately
for line in lines:
    data = line.split()     # split by space
    
    city = data[0]          # first value → city name
    population = float(data[1])   # second value → population
    area = float(data[2])         # third value → area
    
    cities.append(city)
    populations.append(population)
    areas.append(area)

# a) Display details of all cities
print("Details of all cities:")
for i in range(len(cities)):
    print(cities[i], populations[i], areas[i])

# b) Display city names with population > 10 lakhs
print("\nCities with population more than 10 lakhs:")
for i in range(len(cities)):
    if populations[i] > 10:
        print(cities[i])

# c) Display sum of areas of all cities
total_area = sum(areas)
print("\nTotal area of all cities:", total_area)
```

#Question4: Input two values from user where the first line contains N, the number of test 
cases. The next N lines contain the space separated values of a and b. Perform 
integer division and print a/b. Handle exception in case of ZeroDivisionError or 
ValueError. 
Sample input
1 0
2 $
3 1 
Sample Output:
Error Code: integer division or modulo by zero 
Error Code: invalid literal for int() with base 10: '$' 3

```python
#Program to perform integer division with exception handling 
#Take number of test cases from the user
n = int(input("Enter the number of test cases: "))
#Loop for n test cases:
for i in range(n):
    try:
        a, b = input().split()
        #Convert input to integers:
        a = int(a)
        b = int(b)
        #Perform integer division:
        result = a//b
        #print result if no error occurs:
        print(result)
        # Handle division by zero error
    except ZeroDivisionError as e:
        print("Error Code:", e)

    # Handle invalid integer conversion error
    except ValueError as e:
        print("Error Code:", e)




```

Question5. Create multiple suitable exceptions for a file handling program.

```python
# ----- Custom Exceptions -----

# Exception for empty file
class EmptyFileError(Exception):
    pass

# Exception for invalid data in file
class InvalidDataError(Exception):
    pass


# ----- File Handling Program -----

try:
    # Ask user for file name
    filename = input("Enter file name: ")

    # Try opening file
    file = open(filename, "r")

    # Read content
    content = file.read()

    # Check if file is empty
    if content.strip() == "":
        raise EmptyFileError("File is empty")

    # Split lines
    lines = content.splitlines()

    # Assume file should contain only numbers
    numbers = []
    for line in lines:
        if not line.isdigit():
            raise InvalidDataError("File contains non-numeric data")
        numbers.append(int(line))

    # If everything is correct → show results
    print("File opened successfully!")
    print("Numbers in file:", numbers)
    print("Sum of numbers:", sum(numbers))

# File not found exception
except FileNotFoundError:
    print("Error: File not found")

# Custom empty file exception
except EmptyFileError as e:
    print("Error:", e)

# Custom invalid data exception
except InvalidDataError as e:
    print("Error:", e)

# Close file safely
finally:
    try:
        file.close()
        print("File closed.")
    except:
        pass
```
