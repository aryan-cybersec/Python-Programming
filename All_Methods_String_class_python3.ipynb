#  **Formatting**

## `str.format()`
### Description

The `str.format()` method in Python is used to insert values into a string using placeholders `{}`.  
It allows you to build dynamic, readable, and well-structured strings by replacing placeholders with variables or expressions.

```python
name = "Aryan"
age = 19
msg = "My name is {} and I am {} years old.".format(name, age)
print(msg)
```

## `str.format_map()`
### Description

The `str.format_map()` method is used to format strings by replacing placeholders `{}` with values from a mapping object (usually a dictionary).  
Unlike `str.format()`, which accepts positional and keyword arguments, `str.format_map()` takes a single mapping and uses its keys to fill in the placeholders.

```python
person = {"name": "Aryan", "age": 20}
msg = "My name is {name} and I am {age} years old.".format_map(person)
print(msg)
```

# **Searching & Replacing**

## `str.find()`
### Description

The str.find() method is used to search for a substring inside a string.
It returns the lowest index (position) where the substring is found.
If the substring is not found, it returns -1.


```python
text = "Hello Aryan"
position = text.find("Aryan")
print(position)
```

```python
text = "Hello Aryan"
position = text.find("Python")
print(position)
```

## `str.rfind()`
### Description

The str.rfind() method is used to search for a substring inside a string, but unlike str.find(), it returns the highest index (the last occurrence) where the substring is found.
If the substring is not found, it returns -1.


```python
text = "Hello Aryan Aryan"
position = text.rfind("Aryan")
print(position)
```

```python
text = "Hello Aryan"
position = text.rfind("Python")
print(position)
```

## `str.index()`
### Description

The str.index() method is used to search for a substring inside a string and return the lowest index (position) where the substring is found.
Unlike str.find(), if the substring is not found, str.index() raises a ValueError instead of returning -1.



```python
text = "Hello Aryan"
position = text.index("Aryan")
print(position)
```

```python
text = "Hello Aryan"
position = text.index("Python")   # "Python" is not in the string
print(position)
```

## `str.rindex()`
### Description

The str.rindex() method is used to search for a substring inside a string and return the highest index (the last occurrence) where the substring is found.
Unlike str.rfind(), if the substring is not found, str.rindex() raises a ValueError instead of returning -1.


```python
text = "Hello Aryan Aryan"
position = text.rindex("Aryan")
print(position)
```

```python
text = "Hello Aryan"
position = text.rindex("Python")   # "Python" is not in the string
print(position)
```

## `str.startswith()`
### Description

The str.startswith() method is used to check whether a string starts with a specified prefix.
It returns True if the string begins with the given substring, and False otherwise.




```python
text = "Hello Aryan"
print(text.startswith("Hello"))
```

```python
text = "Hello Aryan"
print(text.startswith("hello"))
```

##  `str.endswith()`
### Description

The str.endswith() method is used to check whether a string ends with a specified suffix.
It returns True if the string ends with the given substring, and False otherwise.





```python
text = "Hello Aryan"
print(text.endswith("Aryan"))
```

```python
text = "Hello Aryan"
print(text.endswith("aryan"))
```

##  `str.count()`
### Description

The str.count() method is used to count the number of non-overlapping occurrences of a substring within a string.
It returns an integer representing how many times the substring appears.



```python
text = "Hello Aryan Aryan"
count = text.count("Aryan")
print(count)
```

```python
text = "Hello Aryan"
count = text.count("aryan")   # lowercase 'a'
print(count)
```

##  `str.replace()`
### Description

The str.replace() method is used to create a new string by replacing all occurrences of a specified substring with another substring.
It does not modify the original string (since strings are immutable in Python), but instead returns a new one.



```python
text = "Hello Aryan"
new_text = text.replace("Aryan", "Python")
print(new_text)
```

```python
text = "Hello Aryan"
new_text = text.replace("aryan", "Python")   # lowercase 'a'
print(new_text)
```

# **Splitting & Joining** 

## `str.split()`
### Description

The str.split() method is used to break a string into a list of substrings based on a specified delimiter (separator).
If no delimiter is provided, it splits the string by whitespace (spaces, tabs, newlines).


```python
text = "Hello Aryan how are you"
words = text.split()
print(words)
```

## `str.rsplit()`
### Description

The str.rsplit() method is used to split a string into a list of substrings, just like str.split().
The difference is that rsplit() performs the split from the right side of the string.
It’s especially useful when you want to limit the number of splits (maxsplit) and keep the left part of the string intact.


```python
text = "Hello Aryan how are you"
words = text.rsplit()
print(words)
```

## `str.splitlines()`
### Description

The str.splitlines() method is used to split a string into a list of lines, breaking at line boundaries.
It recognizes different newline characters (\n, \r, \r\n) and returns each line as a separate element in a list.


```python
text = "Hello\nAryan\nWelcome"
lines = text.splitlines()
print(lines)
```

## `str.partition()`
### Description
The str.partition() method is used to split a string into three parts based on the first occurrence of a specified separator.
It always returns a tuple of three elements:
- The part of the string before the separator
- The separator itself
- The part of the string after the separator
If the separator is not found, the method returns:
- The entire string as the first element
- Two empty strings ('', '') as the second and third elements


```python
text = "apple-banana-cherry"
result = text.partition("-")
print(result)
```

## `str.rpartition`
### Description
The str.rpartition() method is used to split a string into three parts, based on the last occurrence of a specified separator.
It always returns a tuple of three elements:
- The part of the string before the separator
- The separator itself
- The part of the string after the separator
If the separator is not found, the method returns:
- Two empty strings ('', '') as the first and second elements
- The entire string as the third element


```python
text = "Hello Aryan Welcome"
result = text.rpartition(" ")
print(result)
```

## `str.join()`
### Description
The str.join() method is used to concatenate (join) elements of an iterable (like a list, tuple, or set) into a single string, with the string acting as the separator between elements.
It’s essentially the reverse of str.split():
- split() breaks a string into a list.
- join() takes a list (or other iterable of strings) and combines it back into a single string.


```python
words = ["Hello", "Aryan", "Welcome"]
sentence = " ".join(words)
print(sentence)
```

# **String Classification**

## `str.isalpha() `

### Description

The str.isalpha() method is used to check whether all characters in a string are alphabetic letters (A–Z or a–z, plus letters from other alphabets).
- Returns True if every character is a letter.
- Returns False if the string contains any non-letter characters (digits, spaces, punctuation, symbols, etc.).


```python
text = "Aryan"
print(text.isalpha())
```

 ## `str.isdecimal() `

### Description
The str.isdecimal() method checks whether all characters in a string are decimal characters (0–9).
- Returns True if every character is a decimal digit.
- Returns False if the string contains any non-decimal characters (letters, spaces, punctuation, fractions, superscripts, etc.).
It is stricter than str.isdigit() and str.isnumeric() because it only recognizes standard base-10 digits (0–9), not other numeric symbols.


```python
text = "12345"
print(text.isdecimal())
```

 ## `str.isdigit() `

### Description
The str.isdigit() method checks whether all characters in a string are digits.
- Returns True if every character is a digit (0–9, plus some other digit characters like superscripts and subscript numbers).
- Returns False if the string contains any non-digit characters (letters, spaces, punctuation, etc.).
It is less strict than str.isdecimal() because it recognizes additional digit characters beyond just 0–9.


```python
text = "12345"
print(text.isdigit())
```

 ## `str.isnumberic() `

### Description
The str.isnumeric() method checks whether all characters in a string are numeric characters.
- Returns True if every character is numeric.
- Returns False if the string contains any non-numeric characters (letters, spaces, punctuation, etc.).
It is the broadest numeric check among isdecimal(), isdigit(), and isnumeric().
- isdecimal() → Only standard digits (0–9).
- isdigit() → Digits + some other digit characters (like superscripts).
- isnumeric() → Even broader: includes digits, superscripts, fractions, Roman numerals, and other numeric symbols.


```python
text = "123Aryan"
print(text.isnumeric())
```

 ## `str.isdentifier() `

### Description
The str.isidentifier() method checks whether a string is a valid Python identifier.
An identifier is the name used for variables, functions, classes, etc. in Python.
For a string to be a valid identifier, it must follow Python’s naming rules:
- ✅ Can contain letters (A–Z, a–z), digits (0–9), and underscores (_).
- ✅ Cannot start with a digit.
- ✅ Cannot contain spaces or special characters (@, #, $, etc.).
- ✅ Cannot be a reserved keyword (like class, def, return).
- Returns True if the string is a valid identifier, otherwise False.




```python
text = "Aryan" 
print(text.isidentifier())
```

 ## `str.islower() `

### Description
The str.islower() method checks whether all cased characters in a string are lowercase.
- Returns True if every letter in the string is lowercase.
- Returns False if there is at least one uppercase letter.
- Non-alphabetic characters (digits, spaces, punctuation) are ignored in the check — they don’t affect the result.
- An empty string returns False.


```python
text = "aryan"
print(text.islower())
```

```python
text = "Aryan"
print(text.islower())
```

 ## `str.isupper() `

### Description
The str.isupper() method checks whether all cased characters in a string are uppercase.
- Returns True if every letter in the string is uppercase.
- Returns False if there is at least one lowercase letter.
- Non-alphabetic characters (digits, spaces, punctuation) are ignored in the check — they don’t affect the result.
- An empty string always returns False.


```python
text = "ARYAN"
print(text.isupper())
```

```python
text = "aryan"
print(text.isupper())
```

 ## `str.istitle() `

### Description
The str.istitle() method checks whether a string is in title case.
- Title case means:
- Each word starts with an uppercase letter.
- The remaining letters in each word are lowercase.
- Returns True if the string follows this pattern.
- Returns False if any word does not follow title case rules.
- Non-alphabetic characters (digits, spaces, punctuation) are ignored in the check — they don’t affect the result.
- An empty string returns False.


```python
text = "Hello Aryan"
print(text.istitle())
```

```python
text = "HELLO ARYAN"
print(text.istitle())
```

 ## `str.isspace() `

### Description
The str.isspace() method checks whether all characters in a string are whitespace characters.
- Returns True if every character is whitespace.
- Returns False if the string contains any non-whitespace characters (letters, digits, punctuation, etc.).
- Whitespace characters include:
- Space (" ")
- Tab ("\t")
- Newline ("\n")
- Carriage return ("\r")
- Form feed ("\f")
- Vertical tab ("\v")
- An empty string always returns False.


```python
text = "   "
print(text.isspace())
```

```python
text = "\t\n"
print(text.isspace())
```

```python
text = " Aryan "
print(text.isspace())
```

```python
text = ""
print(text.isspace())
```

```python
text = "   !   "
print(text.isspace())
```

 ## `str.isprintable() `

### Description
The str.isprintable() method checks whether all characters in a string are printable.
- Returns True if every character is printable.
- Returns False if the string contains any non-printable characters (like control characters: newline \n, tab \t, carriage return \r, etc.).
- Printable characters include:
- Letters (A–Z, a–z)
- Digits (0–9)
- Punctuation (!, ?, . etc.)
- Whitespace (space " ")
- An empty string returns True (since there are no non-printable characters).


```python
text = "Hello Aryan!"
print(text.isprintable())
```

```python
text = "Hello\nAryan"
print(text.isprintable())
```

```python
text = "Hello\tAryan"
print(text.isprintable())
```

```python
text = ""
print(text.isprintable())
```

```python
text = "123 !@#"
print(text.isprintable())
```
