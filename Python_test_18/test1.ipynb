```python
#Answer 1
def top_k(items, k):
    
    return sorted(items, key=lambda x: -x[1])[:k]


items = [("A", 90), ("B", 95), ("C", 95), ("D", 85)]
k = 2
print(top_k(items, k))  
```

```python
# Answer 3
def missing_number(nums):
    n = len(nums)
    total = n * (n + 1) // 2
    return  total - sum(nums)

nums=[3,0,1]
print(missing_number)
```

```python
# Answer 4
def add_unique(x, seen=None):
   
    if seen is None:
        seen = set()
    if x not in seen:
        seen.add(x)
    return list(seen)

print(add_unique(1))  
print(add_unique(2))  
```
