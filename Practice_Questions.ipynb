```python
def dot_product_matrix(A, B):
    # Check condition: columns of A == rows of B
    if len(A[0]) != len(B):
        raise ValueError("Matrix multiplication not possible: columns of A must equal rows of B")
    
    # Initialize result matrix with zeros
    result = [[0 for _ in range(len(B[0]))] for _ in range(len(A))]
    
    # Perform multiplication
    for i in range(len(A)):          # iterate rows of A
        for j in range(len(B[0])):   # iterate columns of B
            for k in range(len(B)):  # iterate over common dimension
                result[i][j] += A[i][k] * B[k][j]
    
    return result


# Example usage
A = [[1, 2, 3],
     [4, 5, 6]]

B = [[7, 8],
     [9, 10],
     [11, 12]]

print(dot_product_matrix(A, B))
# Output: [[58, 64], [139, 154]]
```

```python
def transpose(matrix):
    # Number of rows = len(matrix)
    # Number of columns = len(matrix[0])
    rows = len(matrix)
    cols = len(matrix[0])
    
    # Create new matrix with swapped dimensions
    result = [[0 for _ in range(rows)] for _ in range(cols)]
    
    # Fill result with transposed values
    for i in range(rows):
        for j in range(cols):
            result[j][i] = matrix[i][j]
    
    return result


# Example usage
A = [[1, 2, 3],
     [4, 5, 6]]

print("Original Matrix:", A)
print("Transpose:", transpose(A))
```

```python
class toyota:

    

    def set_input_value(self,a,b,c):
        self.a = a 
        self.b = b 
        self.c = c
        
    
    
    def dummy_mileage_kmpl(self, engine_l: float, weight_kg: float, tyre_in: float) -> float:
        self.engine_l = engine_l
        self.weight_kg = weight_kg
        self.tyre_in = tyre_in
        mileage = 28 - 4 * engine_l - 0.006 * (weight_kg - 1000) - 0.7 * (tyre_in - 15)
        return max(5.0, mileage)  # clamp to minimum 5 km/L
    
    def city_highway_kmpl(self, engine_l: float, weight_kg: float, tyre_in: float) -> tuple[float, float, float]:
        """
        Returns (base, city, highway) mileage in km/L using dummy multipliers.
        """
        base = self.dummy_mileage_kmpl(engine_l, weight_kg, tyre_in)
        city = 0.85 * base
        highway = 1.10 * base
        return base, city, highway



        

```
