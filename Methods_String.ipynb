*  Return a copy of the string with all the cased characters converted to lowercas"

```python
text="Hello World"
print(text.lower())  # hello world
```

* Return a copy of the string with all the cased characters converted to uppercase.

```python
text="Hello World"
print(text.upper())  # HELLO WORLD
```

```python
text="hello world"
print(text.title())  # Hello World
```

* The capitalize() Return a copy of the string with its first character capitalized and the rest lowercased.


```python
text="hello world"
print(text.capitalize())  # Hello world
```

* The swapcase() method returns a string where all the upper case letters are lower case and vice versa.


```python
text="Hello World"
print(text.swapcase())  # hELLO wORLD
```

*  Casefolding is similar to lowercasing but more aggressive because it is intended to remove all case distinctions in a string. For example, the German lowercase letter 'ß' is equivalent to "ss". Since it is already lowercase, lower() would do nothing to 'ß'; casefold() converts it to "ss"



```python
text="Stra$e"
print(text.casefold())  # stra$e
```

* Write a program that reads a paragraph and counts:
- How many words are all uppercase (isupper()).
- How many words are all lowercase (islower()).
- How many words are in title case (istitle()).


```python
paragraph="Python 3 is the latest, actively supported major version of the general-purpose, high-level, interpreted programming language known for its emphasis on code readability and support for various programming paradigms. It was released to modernize the language and fix inconsistencies in the legacy Python 2.x versions, which are no longer supported. "
words=paragraph.split()

uppercase_count=0
lowercase_count=0
titlecase_count=0

for word in words:
    if word.isupper():
        uppercase_count+=1
    elif word.islower():
        lowercase_count+=1
    elif word.istitle():
        titlecase_count+=1
print("All Uppercase words:")
uppercase_count
print("All Lowercase words:")
lowercase_count
print("All Titlecase words:")
titlecase_count
```

* write a program for Armstrong number

```python
num=153
sum_of_powers=0
temp=num
digits=len(str(num))

while temp>0:
    digit=temp%10
    sum_of_powers+=digit**digits
    temp//=10
if num==sum_of_powers:
    print("Entered Number is An Armstrong Number")
else:
    print("Entered Number is Not An Armstrong Number")
```

* write a program for Palindrome number

```python
num=121
temp=num
rev=0

while temp>0:
    digit=temp%10
    rev=rev*10+digit
    temp//=10
if num==rev:
    print("Entered Number is a Palindrome")
else:
    print("Entered Number is Not a Palindrome")
```
